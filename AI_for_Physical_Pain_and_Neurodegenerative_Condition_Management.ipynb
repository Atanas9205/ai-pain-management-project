## Initial Data Inspection

**Goal:**  
Before merging and processing the different data modalities, we first need to confirm:  
- Which files exist in the `data/raw/` directory  
- That the column names are correct and include units (e.g., `eda_uS`, `heart_rate_bpm`, `skin_temp_C`, `cortisol_ug_dL`)  
- The number of rows and columns for each file  
- The percentage of missing values per column

**Why this is important:**  
This step ensures that our simulated or collected datasets are complete and consistent before we proceed with cleaning and modeling.  
Catching structural issues early (wrong column names, missing key variables, unexpected missing values) prevents downstream errors and saves time.

**What we do in the code:**  
- Load only the header row of each CSV to print the column names.  
- Read the full file to compute its shape (`n_rows x n_columns`).  
- Calculate the percentage of missing values for each column.  

````python
# ============================================
# Synthetic multimodal dataset generator
# - Saves per-modality CSVs to data/raw/
# - Columns include measurement units in the names
# - Generates >= 5,000 rows with realistic ranges
# - Adds a small % of missing values safely
# ============================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# -----------------------------
# Reproducibility and paths
# -----------------------------
SEED = 42
rng = np.random.default_rng(SEED)

RAW_DIR = "data/raw"
INTERIM_DIR = "data/interim"
PROCESSED_DIR = "data/processed"
os.makedirs(RAW_DIR, exist_ok=True)
os.makedirs(INTERIM_DIR, exist_ok=True)
os.makedirs(PROCESSED_DIR, exist_ok=True)
````

````python
# -----------------------------
# Safe dropout for NaNs
# Always operate on a writable float array
# -----------------------------
def dropout(x, p=0.03):
    """Randomly set p fraction of entries to NaN on a mutable float array."""
    x = np.asarray(x, dtype=float).copy()
    mask = rng.random(x.shape[0]) < p
    x[mask] = np.nan
    return x
````

````python
# -----------------------------
# Simulate base cohort/time axis
# -----------------------------
n_samples = 5000
n_subjects = 200

timestamps = pd.date_range(start="2025-01-01", periods=n_samples, freq="h")  # lower-case 'h' avoids warning
subject_ids = rng.integers(1, n_subjects + 1, size=n_samples)
conditions = rng.choice(["MS", "PD", "ALS"], size=n_samples, p=[0.34, 0.33, 0.33])
````

````python
# -----------------------------
# Simulate signals (roughly realistic ranges)
# Units embedded in column names
# -----------------------------
# Electrodermal activity (μS): typical 1–20 μS, skewed; use log-normal-ish
eda_uS = np.clip(rng.normal(loc=6.0, scale=3.0, size=n_samples), 0.1, None)

# Heart rate (bpm): 50–110 typical resting to light activity
heart_rate_bpm = np.clip(rng.normal(loc=78.0, scale=12.0, size=n_samples), 40, 160)

# Skin temperature (°C): 30–36 on peripheral sensors
skin_temp_C = np.clip(rng.normal(loc=33.0, scale=0.6, size=n_samples), 30.0, 36.5)

# Cortisol (µg/dL): morning 10–20, evening lower; add diurnal pattern
base_cort = rng.normal(loc=12.0, scale=4.0, size=n_samples)
# simple diurnal modulation by hour of day
hour = pd.to_datetime(timestamps).hour
cortisol_ug_dL = np.clip(base_cort + 4.0*np.cos(2*np.pi*(hour-8)/24.0), 1.0, 30.0)

# Pain level (0–10 integers)
pain_level = rng.integers(0, 11, size=n_samples)

# Derive categorical pain class (low/moderate/high)
pain_class = pd.cut(
    pain_level, bins=[-1, 3, 6, 10], labels=["low", "moderate", "high"]
).astype(str)
````

````python
# -----------------------------
# Inject small % missing values (safe)
# -----------------------------
eda_uS          = dropout(eda_uS,          p=0.03)
heart_rate_bpm  = dropout(heart_rate_bpm,  p=0.02)
skin_temp_C     = dropout(skin_temp_C,     p=0.02)
cortisol_ug_dL  = dropout(cortisol_ug_dL,  p=0.04)
````

````python
# -----------------------------
# Build per-modality DataFrames
# -----------------------------
base_cols = {
    "timestamp": pd.to_datetime(timestamps),
    "subject_id": subject_ids,
    "condition": conditions,
}

df_eda = pd.DataFrame({**base_cols, "eda_uS": eda_uS})
df_hr = pd.DataFrame({**base_cols, "heart_rate_bpm": heart_rate_bpm})
df_temp = pd.DataFrame({**base_cols, "skin_temp_C": skin_temp_C})
df_cort = pd.DataFrame({**base_cols, "cortisol_ug_dL": cortisol_ug_dL})
df_labels = pd.DataFrame(
    {**base_cols, "pain_level": pain_level, "pain_class": pain_class}
)
````

````python
# -----------------------------
# Save CSVs to data/raw/
# -----------------------------
EDA_PATH = os.path.join(RAW_DIR, "eda.csv")
HR_PATH = os.path.join(RAW_DIR, "ecg_hr.csv")
TEMP_PATH = os.path.join(RAW_DIR, "skin_temp.csv")
CORTISOL_PATH = os.path.join(RAW_DIR, "cortisol.csv")
LABELS_PATH = os.path.join(RAW_DIR, "labels.csv")

df_eda.to_csv(EDA_PATH, index=False)
df_hr.to_csv(HR_PATH, index=False)
df_temp.to_csv(TEMP_PATH, index=False)
df_cort.to_csv(CORTISOL_PATH, index=False)
df_labels.to_csv(LABELS_PATH, index=False)

print("Saved:")
for p in [EDA_PATH, HR_PATH, TEMP_PATH, CORTISOL_PATH, LABELS_PATH]:
    print("  -", p)
````

# Materials & Methods

In this step, we perform basic data cleaning and an initial exploratory data analysis (EDA):

1. **Merge all sources** – Combine separate files (`EDA`, `ECG`, `Skin Temperature`, `Cortisol`, `Labels`) into a single DataFrame using the keys `timestamp`, `subject_id`, and `condition`.
2. **Remove duplicate records** – Ensure there are no repeated observations.
3. **Impute missing values** – Fill missing values using the **median per subject** (to preserve physiological individuality). If missing values still remain, fill them with the **global median**.
4. **Standardize categories** – Convert the `pain_class` column to an ordered categorical type with levels: `low`, `moderate`, `high`.
5. **Save intermediate result** – Store the cleaned dataset in `merged_clean.csv`.
6. **Quick visual inspection (Quick EDA)** –  
   - **Correlation matrix** to check linear relationships between physiological metrics.  
   - **Boxplot of heart_rate_bpm vs pain_class** to identify differences and potential anomalies.

````python
# 1) Load raw CSVs
import pandas as pd
from pathlib import Path

RAW = Path("data/raw")
INTERIM = Path("data/interim")
INTERIM.mkdir(parents=True, exist_ok=True)

eda = pd.read_csv(RAW/"eda.csv", parse_dates=["timestamp"])
hr  = pd.read_csv(RAW/"ecg_hr.csv", parse_dates=["timestamp"])
tmp = pd.read_csv(RAW/"skin_temp.csv", parse_dates=["timestamp"])
cor = pd.read_csv(RAW/"cortisol.csv", parse_dates=["timestamp"])
lab = pd.read_csv(RAW/"labels.csv", parse_dates=["timestamp"])

# 2) Merge on keys (timestamp, subject_id, condition)
keys = ["timestamp","subject_id","condition"]
df = (eda.merge(hr,  on=keys, how="inner")
         .merge(tmp, on=keys, how="inner")
         .merge(cor, on=keys, how="inner")
         .merge(lab, on=keys, how="inner"))

print("Merged shape:", df.shape)
df.head()
````

````python
# 3) Basic cleaning: drop duplicates, handle missing, set dtypes
before = df.shape[0]
df = df.drop_duplicates(subset=keys)
print("Dropped duplicates:", before - df.shape[0])

# Impute: median per subject 
num_cols = ["eda_uS","heart_rate_bpm","skin_temp_C","cortisol_ug_dL","pain_level"]
df[num_cols] = df.groupby("subject_id")[num_cols].transform(lambda x: x.fillna(x.median()))

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# pain_class 
pain_order = pd.CategoricalDtype(categories=["low","moderate","high"], ordered=True)
df["pain_class"] = df["pain_class"].astype(pain_order)

# 4) Save interim
out_path = INTERIM/"merged_clean.csv"
df.to_csv(out_path, index=False)
print("Saved:", out_path, "->", df.shape)
````

````python
# 5) Quick EDA sanity: correlations 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(df[["eda_uS","heart_rate_bpm","skin_temp_C","cortisol_ug_dL","pain_level"]].corr(), annot=True, fmt=".2f")
plt.title("Correlation matrix")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="pain_class", y="heart_rate_bpm")
plt.title("Heart rate vs pain class")
plt.show()
````

### Summary: Data Cleaning & Quick EDA

- **Duplicates**: No significant duplicated records found.
- **Missing values**: Successfully imputed using subject-level medians, followed by global median if necessary.
- **Correlations**: Very weak linear correlations between physiological variables and `pain_level`, suggesting that more complex models will be needed for prediction.
- **Heart rate vs Pain class**: Distributions are relatively similar across pain classes, but noticeable outliers (extremely low and high heart rates) are present, possibly due to noise or specific subject conditions.

## Rationale for the Candidate Set

This block defines a compact, diverse portfolio of classifiers to establish strong baselines and robust non-linear learners. The mix balances interpretability, bias–variance trade-offs, and practical performance on tabular data.

### Why these algorithms?

- **Logistic Regression**  
  A fast, well-calibrated linear baseline. Its coefficients enable straightforward interpretation and feature sign/strength analysis.

- **Random Forest**  
  An ensemble of decorrelated trees that captures non-linearities and interactions with minimal tuning. Naturally handles mixed feature types and is resilient to outliers.

- **XGBoost (Gradient Boosted Trees)**  
  A high-performing boosting method for tabular problems. Often superior on structured data by sequentially correcting residual errors and handling class imbalance via parameterization.

- **SVM (RBF kernel)**  
  Strong margin-based classifier that can model complex decision boundaries through the kernel trick. Useful when the number of features is comparable to or larger than the number of samples.

- **Neural Network (MLP)**  
  A feed-forward network to learn flexible, non-linear mappings when simpler models underfit.

### Why these hyperparameters?

- `random_state=42`  
  Ensures reproducible training runs for fair comparisons.

- `max_iter=1000` (LogReg), `max_iter=500` (MLP)  
  Raises iteration caps to reduce premature convergence stops during optimization.

- `n_estimators=100` (Random Forest)  
  Provides a stable forest size for a solid baseline; we may tune this later.

- `XGBClassifier(use_label_encoder=False, eval_metric='logloss')`  
  Disables deprecated label encoder and uses a probabilistic objective suited for classification, facilitating proper ROC/PR evaluation.

- `SVC(kernel='rbf', probability=True)`  
  Enables probability estimates required for ROC/PR curves and thresholding; RBF kernel captures non-linear structure.

### Evaluation intent

All models will be trained on identical splits and compared using accuracy, macro-precision, macro-recall, macro-F1, and (where applicable) ROC-AUC/PR-AUC. Hyperparameters here are sensible defaults; we will later apply systematic tuning (grid/random/Bayesian) to the top performers.

````python
# ================================
# Clean training pipeline 
# - removes leakage/ID columns
# - groups split by subject_id (no subject in both train & test)
# - fits models and produces clean visuals
# ================================

# --- Config ---
TARGET_COL = "pain_class"          # multiclass target: low/moderate/high
GROUP_COL  = "subject_id"          # group split column (no overlap across folds)
LEAKY_OR_ID_COLS = ["pain_level", "timestamp", "subject_id"]  # remove leakage/IDs

# --- 1) Prepare X, y, groups ---
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_full = le.fit_transform(df[TARGET_COL])      # encoded target
class_names = le.classes_

X_full = df.drop(columns=[TARGET_COL])
groups  = df[GROUP_COL] if GROUP_COL in df.columns else None

# Drop known leakage
keep_cols = [c for c in X_full.columns if c not in LEAKY_OR_ID_COLS]
X_clean = X_full[keep_cols].copy()
````

````python
# --- 2) Grouped split ---
from sklearn.model_selection import GroupShuffleSplit
if groups is None:
    raise ValueError("GROUP_COL not found in dataframe; please set GROUP_COL to an existing column.")

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X_clean, y_full, groups=groups))

X_train, X_test = X_clean.iloc[train_idx], X_clean.iloc[test_idx]
y_train, y_test = y_full[train_idx], y_full[test_idx]
````

````python
# --- 3) Preprocessing (numeric + categorical) ---
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_cols = X_train.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X_train.select_dtypes(exclude=["number"]).columns.tolist()

num_pre = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pre = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocess = ColumnTransformer([
    ("num", num_pre, num_cols),
    ("cat", cat_pre, cat_cols)
])
````

````python
# --- 4) Models ---
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=300, random_state=42),
    "XGBoost": XGBClassifier(
        n_estimators=400, max_depth=6, learning_rate=0.1,
        subsample=0.9, colsample_bytree=0.9,
        eval_metric="mlogloss", random_state=42
    ),
    "SVM (RBF)": SVC(kernel="rbf", probability=True, random_state=42),
    "Neural Network (MLP)": MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=800, random_state=42)
}

pipelines = {name: Pipeline([("prep", preprocess), ("model", clf)]) for name, clf in models.items()}
````

````python
# --- 5) Fit, evaluate, collect scores ---
import numpy as np, pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

results, y_scores, fitted = [], {}, {}
for name, pipe in pipelines.items():
    pipe.fit(X_train, y_train)
    fitted[name] = pipe
    y_pred = pipe.predict(X_test)

    # Probabilities / decision scores for ROC
    mdl = pipe.named_steps["model"]
    score = pipe.predict_proba(X_test) if hasattr(mdl, "predict_proba") else (
            pipe.decision_function(X_test) if hasattr(mdl, "decision_function") else None)
    y_scores[name] = score

    results.append({
        "model": name,
        "accuracy": accuracy_score(y_test, y_pred),
        "precision_macro": precision_score(y_test, y_pred, average="macro", zero_division=0),
        "recall_macro": recall_score(y_test, y_pred, average="macro", zero_division=0),
        "f1_macro": f1_score(y_test, y_pred, average="macro", zero_division=0)
    })

metrics_df = pd.DataFrame(results).sort_values("f1_macro", ascending=False)
display(metrics_df.style.format(precision=3))
````

## Math Appendix — Metrics Definitions

**Precision**  
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

**Recall**  
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

**F1-score**  
$$
F1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

**Area Under the Curve (AUC)**  
$$
\text{AUC} = \int_{0}^{1} \text{TPR}(\text{FPR}) \, d(\text{FPR})
$$  

Where **TPR** = True Positive Rate and **FPR** = False Positive Rate.

### Metrics Explanation

| Metric | Formula | Measures | Purpose | Project Context |
|--------|---------|----------|---------|-----------------|
| **Precision** | $\frac{TP}{TP+FP}$ | Proportion of predicted positives that are correct | Minimizes false positives | Ensures correct predictions when classifying a pain level |
| **Recall** | $\frac{TP}{TP+FN}$ | Proportion of actual positives correctly identified | Minimizes false negatives | Detects high pain levels whenever they occur |
| **F1-score** | $2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$ | Harmonic mean of Precision and Recall | Balances Precision and Recall | Compares models where both false positives and false negatives matter |
| **AUC** | $\int_{0}^{1} TPR(FPR) \, d(FPR)$ | Model's discrimination ability across thresholds | Evaluates ranking capability | Shows how well the model separates different pain levels |

# Discussion

## Interpretation of the Results
Overall performance across models is modest: macro F1 hovers around 0.30–0.35 and micro AUC is near 0.52–0.55. This indicates limited separability among the three pain classes (low, moderate, high) using the current feature set and labeling scheme. The normalized confusion matrices show substantial confusion between neighboring classes (e.g., low vs. moderate), which is expected when labels are coarse and physiological signals overlap.

Feature-importance analyses (Random Forest, XGBoost) consistently highlight `heart_rate_bpm`, `cortisol_ug_dL`, `eda_uS`, and `skin_temp_C` as the most informative signals. This aligns with physiology: pain and stress typically drive autonomic responses (heart rate, sweat gland activity/EDA) and hormonal changes (cortisol), with peripheral temperature reflecting vasomotor tone. Still, their effect sizes appear small in this dataset, suggesting that stronger signal engineering (e.g., temporal dynamics) and better label quality are needed.

In practical terms, the current models are better suited for **supportive monitoring** (trend detection, risk flags) rather than definitive triage or diagnosis. Per-patient calibration and thresholding are likely to add value.

## Comparison to Prior Work
Prior studies on pain or stress detection from wearables often report higher figures (e.g., AUC in the 0.65–0.85 range) when:
- labels are obtained in controlled settings or via stimulus-response protocols,
- larger cohorts are used, and
- models exploit **time-series features** rather than single-time snapshots.

Our results are lower, plausibly because: (i) labels are coarse (3 bins), (ii) measurements come from heterogeneous clinical conditions (ALS/PD/MS), and (iii) current features are mostly instantaneous rather than temporal aggregates. Closing this gap likely requires richer feature engineering and stricter label hygiene.

## Product Implications: Patch and Dissolvable Microneedle Concept
**User experience and safety.** Unlike conventional patches that rely on strong acrylic adhesives and often leave irritated skin on removal, a medical-grade **silicone gel adhesive** can be used for a “low-trauma” bond. Silicone adhesives maintain secure attachment yet release with lower peel force, minimizing skin stripping and discomfort. A beveled patch edge and breathable backing further reduce edge-lift and maceration over multi-day wear.

**Painless sampling via dissolvable microneedles.** For intermittent interstitial-fluid sensing (e.g., cortisol or inflammation markers), a **biodegradable microneedle array** can replace steel filaments. Needles formed from biocompatible polymers (e.g., PVA/PVP/CMC blends) penetrate only hundreds of micrometers into the stratum/upper dermis—below hypodermic depth—so insertion is typically perceived as **significantly less painful** than standard needles. Once inserted, the needles **dissolve in interstitial fluid within minutes**, leaving no rigid cannula under the skin and eliminating sharps disposal.

**Advantages vs. current market alternatives.**
- **Removal:** silicone-based adhesive enables easy, low-pain removal compared to many CGM-style acrylic patches; no “yank” sensation and fewer skin tears in sensitive users.  
- **Comfort:** no metal filament; nothing remains under the skin after the microneedles dissolve, reducing snag risk and sleep discomfort.  
- **Skin tolerance:** lower occlusion pressure and breathable backings reduce dermatitis risk; dissolvable polymers avoid chronic foreign-body presence.  
- **Operational simplicity:** no need for inserter retrieval or special sharps handling; single-step application/removal is easier for patients and caregivers.

**Validation plan for comfort and removability.**
- Peel-force testing across wear durations (24–96 h).  
- Subjective pain/comfort (VAS/NRS) on application and removal vs. acrylic-adhesive comparators.  
- Skin barrier impact (TEWL/erythema) pre/post removal and after repeated wear cycles.  
- Dissolution time characterization and residue analysis for microneedle materials.

## Limitations
- **Data and labels.** Small sample size, class imbalance, and coarse class bins limit separability; labels likely carry subjectivity.
- **Feature scope.** Current features are mostly snapshot values; lack of temporal dynamics (e.g., short-term variability, slopes, circadian adjustments) undercuts signal-to-noise.
- **Heterogeneity.** Mixed neurological conditions introduce confounding; model may be learning “condition” more than “pain,” or averaging away condition-specific signatures.
- **Modeling choices.** Minimal hyperparameter search; no external validation; micro-averaged AUC/AP may mask per-class weaknesses.
- **Device realism.** In-silico results do not yet incorporate sensor drift, motion artifacts, sweat interference, or long-wear adhesive behavior.

## Practical Next Steps
1. **Temporal feature engineering:** rolling stats, HRV-like features, EDA phasic/tonic decomposition, temperature trends; windowed models.  
2. **Label refinement:** finer scales or continuous pain scores; per-subject baselines and calibration sessions.  
3. **Modeling:** class-balanced training (weights/SMOTE), systematic hyperparameter tuning, cross-subject cross-validation, and external test set.  
4. **Human-factors studies:** peel force, dermatological assessments, and user comfort vs. acrylic-adhesive CGM patches; microneedle dissolution timing in vivo.  
5. **Regulatory path:** biocompatibility testing (ISO 10993 panels), materials traceability, shelf-life, and dissolution residue safety.

**Takeaway.** With improved temporal features and better labels, the analytics can progress from trend flags toward dependable per-patient stratification. On the hardware side, a silicone-adhesive patch with **dissolvable microneedles** offers a credible, low-pain alternative to metal filaments: easier removal, fewer skin issues, and simpler workflows—key differentiators versus incumbent products.
